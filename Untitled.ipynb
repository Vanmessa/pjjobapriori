```python
import pandas as pd
import sqlite3
```

```python
connect = sqlite3.connect('apriori.sqlite3')
```

```python
ActSol= pd.read_csv('rules2222.csv')
SolDor= pd.read_csv('rules3333.csv')
```

```python
ActSol.to_sql("rulesActSol", connect, if_exists='fail')
SolDor.to_sql("rulesSolDor", connect, if_exists='fail')
```

```python
%load_ext sql
```

```python
%sql sqlite:///apriori.sqlite3
```

```python
%sql SELECT * FROM rulesActSol LIMIT 5
```

```python
%sql SELECT * FROM rulesSolDor LIMIT 5
```

```python
%%sql
CREATE TABLE merge2 AS SELECT rulesActSol.antecedents,rulesSolDor.consequents 
FROM  rulesActSol
INNER JOIN rulesSolDor
ON rulesActSol.consequents = rulesSolDor.antecedents
```

```python
%sql SELECT * FROM merge2
```
